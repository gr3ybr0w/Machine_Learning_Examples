```
import numpy as np
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()

incomes = np.random.normal(100.0, 20.0, 1000)
```

```
incomes_df = pd.DataFrame(incomes, columns=['samples'])
```

```
alt.Chart(incomes_df).mark_bar().encode(
    x=alt.X('samples:Q', bin=alt.Bin(maxbins=50)),
    y='count()'
)
```

```
incomes.std()
```

```
incomes.var()
```
