# Percentiles

```python
import numpy as np
import pandas as pd
import altair as alt
alt.data_transformers.disable_max_rows() # disable record limit for altair

# make normaly distributed array
vals = np.random.normal(0, 0.5, 10000)

# making DataFrame of the data for use in altair
vals_df = pd.DataFrame(vals, columns=['vals'])
```

```python
alt.Chart(vals_df).mark_bar().encode(x=alt.X('vals:Q',
                                             bin=alt.Bin(maxbins=100)),
                                     y='count()',
                                     tooltip=['count()'])
```

```python
np.percentile(vals, 50)
```

```python
np.percentile(vals, 90)
```

```python
np.percentile(vals, 20)
```
