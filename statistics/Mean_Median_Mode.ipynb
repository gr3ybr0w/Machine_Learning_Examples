# Mean, Median, Mode and introducing Numpy

## Mean vs. Median

Let's create some fake income data, centered around 27,000 with a normal distribution and standard deviation of 15,000 with 10,000 data points. (We'll dicuss those terms more later, if you're not famliar with them.)


Then, compute the mean (average) - it should be close to 27,000:


```
import altair as alt
import numpy as np
import pandas as pd

alt.data_transformers.disable_max_rows()
```

```
incomes = np.random.normal(27000, 15000, 10000)
np.mean(incomes)
```

```
incomes_df = pd.DataFrame(incomes, columns=['samples'])
```

```
alt.Chart(income_df).mark_bar().encode(
    x=alt.X('samples:Q', bin=alt.Bin(maxbins=50)),
    y='count()',
)
```

```
np.median(incomes)
```

```
incomes = np.append(incomes, [10000000000])
```

```
np.mean(incomes)
```

```
np.median(incomes)
```
